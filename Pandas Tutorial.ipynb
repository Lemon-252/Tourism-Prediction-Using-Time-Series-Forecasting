```python
#How to remove duplicate rows
import pandas as pd
user_cols=['user_id', 'age','gender','occupation','zip_code']
users=pd.read_table('http://bit.ly/movieusers',sep='|',header=None,names=user_cols)
users.set_index('user_id',inplace=True)
users.head()
users.duplicated()#returns a boolean series
users.loc[users.duplicated(keep='first'),:]#to see the duplicated entries
users.loc[users.duplicated(keep='last'),:]#to see the duplicated entries
users.loc[users.duplicated(keep=False),:]#to see the duplicated entries
```

```python
#to drop duplicate rows use the following 
users.drop_duplicates(keep=False).shape
```

```python
#what if we only want to drop certain columns
users.loc[users.duplicated(subset=['age','zip_code'],keep=False),:]#this line of code will consider duplicacy on the basis of specified columns:age and zip code
users.duplicated(subset=['age','zip_code'],keep=False).sum()#32 such items present
users.drop_duplicates(subset=['age','zip_code'],keep=False).shape#32 items dropped
```

```python
#map,apply and applymap
import pandas as pd
train=pd.read_csv('http://bit.ly/kaggletrain')
train.head()
```

```python
train['sexnum']=train.Sex.map({'female':0,'male':1})#map is a series method
train.loc[0:4,['Sex','sexnum']]
#apply is a series as well as dataframe method
#applies a function to each element in a series
train['Name_length']=train.Name.apply(len)
train.loc[0:4,['Name','Name_length']]
```

```python
import numpy as np
train['Fare_ceil']=train.Fare.apply(np.ceil)
```

```python
train.loc[0:4,['Fare','Fare_ceil']]
```

```python
#how to get the sirnames of the passenger list
train.Name.str.split(',').head()
def get_element(mylist,position):
    return mylist[position]
train.Name.str.split(',').apply(get_element,position=0).head()
```
